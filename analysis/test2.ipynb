```python
import numpy as np
from scipy.optimize import minimize

# Function to calculate model torque based on voltage input
def model_torque_U(params, df):
    k_e, R = params
    k_t = 2.6657
    df['tau_U'] = (k_t / R) * (df['U'] - k_e * df['DXL_Velocity'])
    return df['tau_U']

# Function to calculate model torque based on current input
def model_torque_I(params, df):
    k_e, R = params  # k_e is not used in this function but included for consistency
    k_t = 2.6657
    df['tau_I'] = df['DXL_Current'] * k_t
    return df['tau_I']

# Objective function to minimize
def objective(params, df):
    tau_U_pred = model_torque_U(params, df)
    tau_I_pred = model_torque_I(params, df)
    error = np.sum((tau_U_pred - tau_I_pred) ** 2)
    return error
# 4 Ohm, 0.072 
# Initial guess for the parameters [k_e, R]
initial_guess = [0.1, 1]

# Run the optimization
result = minimize(fun=objective, x0=initial_guess, args=(df,), method='L-BFGS-B', bounds=[(0, None), (0, None)])

if result.success:
    optimized_ke, optimized_R = result.x
    print(f"Optimized k_e: {optimized_ke}, Optimized R: {optimized_R}")
else:
    print("Optimization failed:", result.message)
```
