```python
import numpy as np
import pandas as pd

# Constants
time_step = 5e-3  # Time step in seconds
total_time = 120  # Total time for the experiment
time = np.arange(0, total_time, time_step)
fundamental_period = 12  # Fundamental period for both waves
omega = 2 * np.pi / fundamental_period  # Angular frequency of the fundamental frequency
amplitude = 0.14 # Amplitude of the waves


# Function to generate Fourier series for square wave up to the 5th harmonic
def generate_fourier_square_wave(time, omega, amplitude, n_harmonics=5):
    wave = np.zeros_like(time)
    for n in range(1, n_harmonics * 2, 2):  # Only odd harmonics
        wave += (4 * amplitude / (n * np.pi)) * np.sin(n * omega * time)
    return wave

# Function to generate Fourier series for sawtooth wave up to the 5th harmonic
def generate_fourier_sawtooth_wave(time, omega, amplitude, n_harmonics=5):
    wave = np.zeros_like(time)
    for n in range(1, n_harmonics + 1):  # Both odd and even harmonics
        wave += (2 * amplitude / (n * np.pi)) * (-1) ** (n + 1) * np.sin(n * omega * time)
    return wave

# Function to generate Fourier series for triangle wave up to the 5th harmonic
def generate_fourier_triangle_wave(time, omega, amplitude, n_harmonics=5):
    wave = np.zeros_like(time)
    for n in range(1, n_harmonics * 2, 2):  # Only odd harmonics
        wave += ((8 * amplitude) / ((n * np.pi) ** 2)) * np.sin(n * omega * time)
    return wave

# Generate the waves
square_wave = generate_fourier_square_wave(time, omega, amplitude)
sawtooth_wave = generate_fourier_sawtooth_wave(time, omega, amplitude)
triangle_wave = generate_fourier_triangle_wave(time, omega, amplitude*1.5)

units = 1000/2.69

# Create DataFrames
df_square = pd.DataFrame({
    'time': time,
    'q1_l': square_wave * units,
    'q1_r': square_wave * units,
    'q2_l': square_wave * units,
    'q2_r': square_wave * units
})

df_sawtooth = pd.DataFrame({
    'time': time,
    'q1_l': sawtooth_wave * units,
    'q1_r': sawtooth_wave * units,
    'q2_l': sawtooth_wave * units,
    'q2_r': sawtooth_wave * units
})

# Create DataFrame
df_triangle = pd.DataFrame({
    'time': time,
    'q1_l': triangle_wave * units,
    'q1_r': triangle_wave * units,
    'q2_l': triangle_wave * units,
    'q2_r': triangle_wave * units
})

# Save to CSV files
df_square.to_csv(f'square_wave_ref_{fundamental_period}.csv', index=False)
df_sawtooth.to_csv(f'sawtooth_wave_ref_{fundamental_period}.csv', index=False)
df_triangle.to_csv(f'triangle_wave_ref_{fundamental_period}.csv', index=False)
```
